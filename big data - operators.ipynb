```python
import pyspark
import pandas as pd
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col , format_number

spark = SparkSession.builder.appName('Sales').getOrCreate()
```

```python
data = spark.read.csv("/content/Datasets.xlsx - Dataset-Sales.csv", header=True, inferSchema=True)
data.show()
```

```python
data.printSchema()
```

1.filter operation - extract rows from DF based on conditions.

```python
from pyspark.sql.functions import col

data.filter(col("Total Sales") > "$90000").show()
```

2.logical operators
* AND operator - &
* OR operator - |

```python
data.filter((col("Total Sales") > "$10000") & (col("Total Sales") < "$20000")).show()
```

```python
data.filter((col("Total Sales") > "$10000") | (col("Total Sales") < "$20000")).show()
```

3. Comparison operators ==,!=

```python
data.filter(col("Region") == "West").show()
```

```python
data.filter(col("Region") != "West").show()
```

4. Groupby

```python
from pyspark.sql.functions import col, regexp_replace

data = data \
.withColumn("Price per Unit", regexp_replace(col("Price per Unit"), "[$, ]", "").cast("double")) \
.withColumn("Units Sold", regexp_replace(col("Units Sold"), ",", "").cast("int")) \
.withColumn("Total Sales", regexp_replace(col("Total Sales"), "[$, ]", "").cast("double")) \
.withColumn("Operating Profit", regexp_replace(col("Operating Profit"), "[$, ]", "").cast("double")) \
.withColumn("Operating Margin", regexp_replace(col("Operating Margin"), "%", "").cast("double"))
```

```python
data.printSchema()
```

```python
data.groupBy("Region").sum("Total Sales").show()
```

```python
data.groupBy("Retailer").count().show()
```

5. Overall Aggregated values function

```python
from pyspark.sql.functions import avg, sum, max, min

data.groupBy("Region").agg(
    sum("Total Sales").alias("Total_Sales"),
    avg("Operating Profit").alias("Avg_Profit"),
    max("Total Sales").alias("Max_Sales"),
    min("Total Sales").alias("Min_Sales")
).show()
```


6. format_number, ALias

```python
from pyspark.sql.functions import format_number, avg

data.groupBy("Region").agg(
    format_number(avg("Total Sales"), 2).alias("Avg_Sales")
).show()
```


7. Sort data using orderBy

```python
data.orderBy("Total Sales").show()
```

```python
from pyspark.sql.functions import col

data.orderBy(col("Total Sales").desc()).show()
```
