1. Filter employees with salary greater than 50,000.

##Operations

1. Group by department and count employees in each department.







```python
print("\n6. Employees count by department:")
df.groupBy("department").count().show()
```

2. Find total salary per department.

```python
print("\n7. Total salary per department:")
df.groupBy("department").agg(sum("salary").alias("total_salary")).show()
```

3. Find average salary per department.

```python
print("\n8. Average salary per department:")
df.groupBy("department").agg(avg("salary").alias("avg_salary")).show()
```

4. Group by city and find max salary.

```python
print("\n9. Max salary by city:")
df.groupBy("city").agg(spark_max("salary").alias("max_salary")).show()
```

5. Group by department and gender and count employees.

```python
print("\n10. Employee count by department and gender:")
df.groupBy("department", "gender").count().show()
```

# Multiple Aggregation
1.Perform multiple aggregations on salary per department:
*   . sum
*   . average
*   . max
*   . min



```python
print("\n11.1 SUM salary per department:")
df.groupBy("department").agg(sum("salary").alias("total_salary")).show()
```

```python
print("\n11.2 AVERAGE salary per department:")
df.groupBy("department").agg(avg("salary").alias("avg_salary")).show()
```

```python
print("\n11.3 MAX salary per department:")
df.groupBy("department").agg(spark_max("salary").alias("max_salary")).show()
```

```python
print("\n11.4 MIN salary per department:")
df.groupBy("department").agg(spark_min("salary").alias("min_salary")).show()
```

Group by department and role, then calculate:


```python

print("\n11.5 ALL aggregations together:")
df.groupBy("department").agg(
    sum("salary").alias("total_salary"),
    avg("salary").alias("avg_salary"),
    spark_max("salary").alias("max_salary"),
    spark_min("salary").alias("min_salary")
).show()
```

Total salary


```python
print("\n12.1 Total salary by department and role:")
df.groupBy("department", "role").agg(sum("salary").alias("total_salary")).show()
```

Average salary


```python

print("\n12.2 Average salary by department and role:")
df.groupBy("department", "role").agg(avg("salary").alias("avg_salary")).show()
```

Employee count

```python

print("\n12.3 Employee count by department and role:")
df.groupBy("department", "role").agg(count("*").alias("employee_count")).show()
```

Count distinct roles in each department.

```python

print("\n13. Distinct roles count by department:")
df.groupBy("department").agg(countDistinct("role").alias("distinct_roles")).show()
```


Find the top 3 departments with the highest total salary.

```python

print("\n14. Top 3 departments by total salary:")
df.groupBy("department").agg(sum("salary").alias("total_salary")) \
  .orderBy("total_salary", ascending=False) \
  .limit(3) \
  .show()

print("\nAssignment completed successfully before 12:00 a.m!")
spark.stop()
```
